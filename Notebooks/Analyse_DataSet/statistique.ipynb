```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Ajuster les options d'affichage
pd.set_option('display.max_columns', None)  # Aucune limite sur le nombre de colonnes affich√©es
pd.set_option('display.max_colwidth', None)  # Aucune limite sur la largeur du contenu de la colonne
```

```python
# Correctly specifying your file path
path_global_5s = "../../Dataset acoustique insectes/CSVs morceaux audio 5s/Global/train_global_recording_chunks.csv"

# Using pandas to read the CSV file into a DataFrame
global_5s = pd.read_csv(path_global_5s)
```

```python
global_5s.iloc[0].to_frame().T
```

```python
# Statistiques Descriptives
global_5s.describe(include='all')
```

```python
# Informations sur les Types de Donn√©es
global_5s.info()
```

```python
# Comptage des Valeurs Uniques
for col in global_5s.columns:
    print(f"{col}: {global_5s[col].nunique()} unique values")
```

```python
# Valeurs Manquantes
global_5s.isnull().sum()
```

```python
# Corr√©lations
corr = global_5s.corr()

# Cr√©er une heatmap pour visualiser la matrice de corr√©lation
plt.figure(figsize=(20, 16))  # Ajuste la taille de la figure (vous pouvez modifier ces valeurs selon vos besoins)
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True, linewidths=.5)

# Titre et ajustements optionnels
plt.title('Matrice de corr√©lation pour global_5s')
plt.xticks(rotation=45, ha='right')  # Rotation des √©tiquettes de l'axe x pour une meilleure lisibilit√©
plt.yticks(rotation=45)  # Rotation des √©tiquettes de l'axe y
plt.tight_layout()  # Ajuste automatiquement les param√®tres de la subplot pour qu'elle s'adapte √† la zone de la figure

# Afficher la figure
plt.show()
```

```python
# Compter les Occurrences de Chaque Label
global_5s['label'].value_counts()
```

```python
# Calculer des Statistiques de Base pour les Colonnes Num√©riques par Label
global_5s.groupby('label').describe()
```

```python
# Visualiser la Distribution des Labels

plt.figure(figsize=(30, 12))
sns.countplot(x='label', data=global_5s)
plt.title('Distribution des Labels')
plt.xticks(rotation=45)
plt.show()
```

```python
# Liste des colonnes num√©riques que vous souhaitez explorer
colonnes_numeriques = ['min_frequency', 'max_frequency', 'duree_min', 'duree_sec']

# Boucle pour cr√©er un boxplot pour chaque variable num√©rique
for colonne in colonnes_numeriques:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='label', y=colonne, data=global_5s)
    plt.title(f'Boxplot de {colonne} par Label')
    plt.xticks(rotation=45)
    plt.show()
```
