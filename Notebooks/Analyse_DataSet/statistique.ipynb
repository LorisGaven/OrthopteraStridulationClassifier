```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = global_5s

# Cr√©ation d'une nouvelle colonne pour les positions x uniques des labels
# Cela est n√©cessaire pour g√©rer les labels r√©p√©t√©s
unique_labels, label_positions = np.unique(df['label'], return_inverse=True)
df['label_position'] = label_positions

# Cr√©ation du graphique
plt.figure(figsize=(10, 6))  # Ajustez la taille selon vos besoins

# Boucle sur les lignes du DataFrame pour tracer chaque segment de ligne
for _, row in tqdm(df.iterrows()):
    plt.plot([row['label_position'], row['label_position']], [row['min_frequency'], row['max_frequency']])
    
# Ajustement des param√®tres du graphique
plt.xticks(range(len(unique_labels)), unique_labels)  # D√©finit les labels sur l'axe des x
plt.xlabel('Label')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.title('Frequency Range by Label')
plt.grid(True, axis='y')  # Affiche une grille horizontale pour une meilleure lisibilit√©

# Affichage du graphique
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np


for i in range(0,56,5):
    # Charger le fichier audio
    file = f'Anura_cdbomar12FR_27IV2022_2145_split_{i}_{i+5}'
    filename = "../../Dataset acoustique insectes/S√©lection morceaux audio 5s/Audible/train/" + file + ".wav" # Remplacez 'your_file.wav' par le chemin vers votre fichier
    filtered_df = df[(df['code_unique'] == "Anura_cdbomar12FR_27IV2022_2145")
                & (df['chunk_initial_time'] == i)]
    
    try:
        y, sr = librosa.load(filename)
    except:
        continue

    # Calculer le spectrogramme de puissance
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

    # Afficher le spectrogramme
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')

    for index, row in filtered_df.iterrows():
        # Calculer les coordonn√©es x de la bounding box
        x_start = row["annotation_initial_time"] - row["chunk_initial_time"]
        x_end = row["chunk_final_time"] - row["annotation_initial_time"]
        width = x_end - x_start
        
        # Coordonn√©es y et hauteur de la bounding box
        y_start = row["min_frequency"]
        height = row["max_frequency"] - y_start
        
        # Cr√©er et ajouter la bounding box au graphique
        rect = patches.Rectangle((x_start, y_start), width, height, linewidth=1, edgecolor='r', facecolor='none')
        plt.gca().add_patch(rect)

    plt.colorbar(format='%+2.0f dB')
    plt.title('Power spectrogram')
    plt.tight_layout()
    plt.show()
```
